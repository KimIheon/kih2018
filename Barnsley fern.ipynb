```python
import numpy as np
import matplotlib.pyplot as plt
import random

plt.figure(figsize = (4, 5), dpi = 100)

start = np.array([[0], [0]])
next_product1 = np.array([[0,0], [0,0.16]])
next_product2 = np.array([[0.85, 0.04], [-0.04, 0.85]])
next_product3 = np.array([[0.20, -0.26], [0.23, 0.22]])
next_product4 = np.array([[-0.15, 0.28], [0.26, 0.24]])

for i in range(10000):
    r = random.randint(1, 100)
    
    if r == 1:
        start = np.dot(next_product1, start)
        plt.plot(start[0], start[1], 'k', marker='.', markersize=1)
    elif r < 87:
        start = np.dot(next_product2, start) + np.array([[0],[1.6]])
        plt.plot(start[0], start[1], '.k', marker='.', markersize=1)
    elif r < 94:
        start = np.dot(next_product3, start) + np.array([[0],[1.6]])
        plt.plot(start[0], start[1], '.k', marker='.', markersize=1)
    else:
        start = np.dot(next_product4, start) + np.array([[0],[0.44]])
        plt.plot(start[0], start[1], '.k', marker='.', markersize=1)

plt.show()
```
