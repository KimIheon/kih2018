```python
import csv

class Graph :
    def __init__(self, burger_file_name, gdp_file_name) :
        self.b_name = burger_file_name
        self.g_name = gdp_file_name
        self.b_data = []
        self.g_data = []
        with open(self.b_name) as file :
            data = csv.reader(file)
            next(data)
            for row in data :
                self.b_data.append(row)
        with open(self.g_name) as file :
            data = csv.reader(file)
            next(data)
            for row in data :
                self.g_data.append(row)
                
    def fix_burger(self) :
        burger_list = []
        for row in self.b_data :
            if '' not in row :
                burger_list.append((float(row[1]) + float(row[2]) + float(row[3]) + float(row[4]))
                                   / (float(row[5]) + 1))
        return burger_list
    
    def city(self) :
        city_list = []
        for row in self.b_data :
            if '' not in row :
                city_list.append(row[0])
        return city_list
    
    def gdp(self) :
        gdp_list = []
        for row in self.g_data :
            gdp_list.append(float(row[1]))
        return gdp_list
    
    def draw_graph(self) :
        plt.figure(figsize = (10, 4), dpi = 200)
        plt.rc('font', family = 'Noto Sans CJK KR')
        plt.plot(self.fix_burger(), self.gdp(), '.')
        plt.title('도시별 수정된 버거지수와 gdp')
        plt.xlabel('x축 : 버거지수')
        plt.ylabel('y축 : gdp')
        for city_name in self.city() :
            _index = self.city().index(city_name)
            plt.annotate(city_name, 
                         xy = (self.fix_burger()[_index], self.gdp()[_index]), xycoords = 'data',
                         xytext = (self.fix_burger()[_index] + 0.05, self.gdp()[_index]), textcoords = 'data', fontsize = 5,
                         arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3"))
        plt.show()
```
